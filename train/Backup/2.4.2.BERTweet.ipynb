```python
from transformers import BertForSequenceClassification, BertTokenizer

# Charger le modèle BERT pour la classification de séquences
bert_model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)  # 2 classes : positif et négatif
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Vos données tokenisées
tabX = []
for sentence in df_sample['words']:
    testX = torch.tensor([tokenizer.encode(sentence)])
    tabX.append(testX)

# Vos embeddings
tabY = []
for i in tabX:
    with torch.no_grad():
        features = bert_model(i)
    tabY.append(features.logits)

# Concaténer les embeddings
features_tensor = torch.cat(tabY, dim=0)

# Calculer les prédictions
predictions = torch.nn.Softmax(dim=-1)(features_tensor)

y_pred = []
# Afficher les résultats
for i, pred in enumerate(predictions):
    predicted_class = torch.argmax(pred).item()
    if predicted_class == 0:
        y_pred.append(0)
        print(f"Sentence {i + 1}: Sentiment négatif")
    else:
        y_pred.append(1)
        print(f"Sentence {i + 1}: Sentiment positif")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(df_sample['target'], y_pred)
print("Accuracy:", accuracy)
```

```python
text = 'I am so happy to see you at this evening!'
input_ids = torch.tensor([tokenizer.encode(text)])
input_ids
```

```python
with torch.no_grad():
    features = bertweet(input_ids)
```

```python
classification_model = torch.nn.Sequential(
    torch.nn.Linear(features.last_hidden_state.shape[-1], 2),  # 2 classes : positif et négatif
    torch.nn.Softmax(dim=-1)
)

# Calculer les prédictions
predictions = classification_model(features.last_hidden_state[:, 0, :])

# Afficher le résultat
predicted_class = torch.argmax(predictions).item()
if predicted_class == 0:
    print("Sentiment négatif")
else:
    print("Sentiment positif")
```
